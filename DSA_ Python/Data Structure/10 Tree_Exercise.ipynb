# Question-1

```python
class Node:
    def __init__(self,name,desig):
        self.name=name
        self.desig=desig
        self.children=[]
        self.parent=None

    def add_child(self,child):
        self.children.append(child)
        child.parent=self

    def level(self):
        lvl=0
        p=self.parent
        while p:
            lvl+=1
            p=p.parent

        return lvl

    def print_tree(self,type):
        space=" "*self.level()*5
        pre=space+"!-->" if self.parent else " "
        if type=="name":
             print(pre,self.name)
             for i in self.children:
                 i.print_tree("name")
        elif type=="designation":
             print(pre,self.desig)
             for i in self.children:
                 i.print_tree("designation")
        elif type=="both":
             print(pre,self.name,"(",self.desig,")")
             for i in self.children:
                 i.print_tree("both")


def build_management_tree():
    CEO=Node("Nilipul","CEO")
    
    CTO=Node("Chinmay","CTO")
    IH=Node("Vishwa","Infastructure Head")
    CM=Node("Dhaval", "Cloud Manager")
    AM=Node("Abhijit","App Manager")

    IH.add_child(CM)
    IH.add_child(AM)

    AH=Node("Aamir","Application Head")
    
    CTO.add_child(IH)
    CTO.add_child(AH)
    
    HR=Node("Gels","HR Head")
    RM=Node("Peter","Recruitment Manager")
    PM=Node("Waqas","Policy Manager")
    
    HR.add_child(RM)
    HR.add_child(PM)
    
    CEO.add_child(CTO)
    CEO.add_child(HR)

    return CEO


if __name__=='__main__':
    root=build_management_tree()
    root.print_tree("name")
    root.print_tree("designation")
    root.print_tree("both")
```

```python
from platform import node


class Node:
    def __init__(self,data):
        self.data=data
        self.children=[]
        self.parent=None

    def add_child(self,child):
        self.children.append(child)
        child.parent=self

    def level(self):
        lvl=0
        p=self.parent
        while p:
            lvl+=1
            p=p.parent

        return lvl

    def print_tree(self,l):
        if self.level()>l :
            return
        space=" "*self.level()*5
        pre=space+"!-->" if self.parent else " "
        print(pre,self.data)
        for i in self.children:
            i.print_tree(l)


def build_tree(): 
    globe=Node("Global")
    
    ind=Node("India")
    guj=Node("Gujarat")
    ahm=Node("Ahmedabad")
    bar=Node("Baroda")

    guj.add_child(ahm)
    guj.add_child(bar)

    kar=Node("Karnataka")
    ban=Node("Bangluru")
    mys=Node("Mysore")

    kar.add_child(ban)
    kar.add_child(mys)
    
    ind.add_child(guj)
    ind.add_child(kar)
    
    usa=Node("USA")
    njer=Node("New Jersey")
    pri=Node("Princeton")
    tre=Node("Trenton")

    njer.add_child(pri)
    njer.add_child(tre)

    cal=Node("California")
    sanf=Node("San Francisco")
    mtv=Node("Mountain View")
    palt=Node("Palo Alto")

    cal.add_child(sanf)
    cal.add_child(mtv)
    cal.add_child(palt)
    
    usa.add_child(njer)
    usa.add_child(cal)
    
    globe.add_child(ind)
    globe.add_child(usa)

    return globe


if __name__=='__main__':
    root=build_tree()
    root.print_tree(0)
    root.print_tree(1)
    root.print_tree(2)
    root.print_tree(3)
```
