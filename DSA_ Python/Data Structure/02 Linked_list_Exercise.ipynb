# Question-1(Singly Linked List)
In Linkedlist class  

```python
from logging import exception
from traceback import print_list


class Node:
    def __init__(self,data=None,link=None):
        self.data=data
        self.link=link

class Linkedlist:
    def __init__(self):
        self.head=None 

    def print_list(self):
        if self.head==None:
            print("Linked List is Empty")
            return
        
        temp=self.head
        while temp:
            print(temp.data,"--> ", end="")
            temp=temp.link
        print("\n")

    def LL_length(self):
        count=0
        temp=self.head
        while temp:
            count+=1
            temp=temp.link
        return count

    def insert_at_beg(self,data):
        newnode=Node(data,self.head)
        self.head=newnode

    def insert_at_end(self,data):
        if self.head==None:
            self.head=Node(data,None)
            return
        
        temp=self.head
        while temp.link:
            temp=temp.link

        temp.link=Node(data,None)

    def insert_at(self,pos,data):
        if pos<1 or pos>self.LL_length():
            raise Exception("Invalid Position Entered")

        if pos==1:
            self.insert_at_beg(data)
            return
        
        count=1
        temp=self.head
        while temp:
            if count==pos-1:
                newnode=Node(data,temp.link)
                temp.link=newnode
                break
            temp=temp.link
            count+=1

    def insert_Values(self,data):
        self.head=None
        for i in data:
            self.insert_at_end(i)

    def delete_at(self,pos):
        if pos<1 or pos>self.LL_length():
            raise Exception("Invalid Position Entered")

        if pos==1:
            self.head=self.head.link
            return
        
        count=1
        temp=self.head
        while temp:
            if count==pos-1:
                temp.link=temp.link.link
                break
            temp=temp.link
            count+=1
    
    #Answer

    def insert_after_value(self,data_after,data):
        if self.head==None:
            self.head=Node(data,None)
            return
        
        temp=self.head
        while temp:
            if temp.data==data_after:
                newnode=Node(data,temp.link)
                temp.link=newnode
                break
            temp=temp.link
        else:
            self.insert_at_end(data)

    def delete_by_value(self,data):
        if self.head==None:
            print("list is already empty")
            return

        if self.head.data==data:
            self.head=self.head.link
            return

        temp=self.head
        while temp.link:
            if temp.link.data==data:
                temp.link=temp.link.link
                break
            temp=temp.link
        else:
            print("Valus does not present in Linked List")
        
        

ll=Linkedlist()
ll.insert_after_value("figs","melon")
ll.print_list()
ll.insert_Values(["banana","mango","grapes","orange"])
ll.print_list()
ll.insert_after_value("figs","apple")
ll.print_list()
ll.insert_after_value("mango","guava")
ll.print_list()
ll.delete_by_value("orange")
ll.print_list()
ll.delete_by_value("figs")
ll.print_list()
ll.delete_by_value("banana")
ll.delete_by_value("mango")
ll.delete_by_value("guava")
ll.delete_by_value("grapes")
ll.delete_by_value("apple")
ll.delete_by_value("figs")  
ll.print_list()
```

# Question-2(Doubly Linked List)
Implement Doubly Linked List

```python
class Node:
    def __init__(self,data=None,prev=None,next=None):
        self.data=data
        self.prev=prev
        self.next=next

class Doubly_Linkedlist:
    def __init__(self):
        self.head=None
        self.tail=None 

    def print_forward(self):
        if self.head==None:
            print("Linked List is Empty")
            return
        
        temp=self.head
        while temp:
            print(temp.data,"--> ", end="")
            temp=temp.next
        print("\n")

    def print_backward(self):
        if self.head==None:
            print("Linked List is Empty")
            return
        
        temp=self.tail
        while temp:
            print(temp.data,"--> ", end="")
            temp=temp.prev
        print("\n")

    def DLL_length(self):
        count=0
        temp=self.head
        while temp:
            count+=1
            temp=temp.next
        return count

    def insert_at_beg(self,data):
        newnode=Node(data,None,self.head)
        self.head.prev=newnode
        self.head=newnode

    def insert_at_end(self,data):
        if self.head==None:
            self.head=self.tail=Node(data,None,None)
            return
        
        temp=self.head
        while temp.next:
            temp=temp.next

        temp.next=self.tail=Node(data,temp,None)

    def insert_at(self,pos,data):
        if pos==self.DLL_length()+1:
            self.insert_at_end(data)
            return

        if pos<1 or pos>self.DLL_length():
            raise Exception("Invalid Position Entered")

        if pos==1:
            self.insert_at_beg(data)
            return
        
        count=1
        temp=self.head
        while temp:
            if count==pos-1:
                newnode=Node(data,temp,temp.next)
                temp.next.prev=newnode
                temp.next=newnode
                break
            temp=temp.next
            count+=1

    def insert_Values(self,data):
        self.head=None
        for i in data:
            self.insert_at_end(i)

    def delete_at(self,pos):
        if pos<1 or pos>self.DLL_length():
            raise Exception("Invalid Position Entered")

        if pos==1:
            self.head=self.head.next
            self.head.prev=None
            return
        
        count=1
        temp=self.head
        while temp:
            if count==pos-1:
                temp.next=temp.next.next
                temp.next.prev=temp
                break
            temp=temp.next
            count+=1

dll=Doubly_Linkedlist()
dll.DLL_length()
dll.insert_Values(["banana","mango","grapes","orange"])
dll.print_forward()
dll.print_backward()
dll.insert_at_end("figs")
dll.print_forward()
dll.print_backward()
dll.insert_at(1,"jackfruit")
dll.print_forward()
dll.print_backward()
dll.DLL_length()
dll.insert_at(7,"dates")
dll.print_forward()
dll.print_backward()
dll.insert_at(3,"kiwi")
dll.print_forward()
dll.print_backward()
dll.delete_at(7)
dll.print_forward()
dll.print_backward()
dll.delete_at(3)
dll.print_forward()
dll.print_backward()
```
