```python
class Binary_Search_Tree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

    def add(self,data):
        if self.data==data:
            return
        
        if self.data>data:
            if self.left:
                self.left.add(data)
            else:
                self.left=Binary_Search_Tree(data)

        else:
            if self.right:
                self.right.add(data)
            else:
                self.right=Binary_Search_Tree(data)

    def search(self,value):
        if self.data==value:
            print("True")

        if self.data>value:
            if self.left:
                self.left.search(value)
            else:
                print("False")

        if self.data<value:
            if self.right:
                self.right.search(value)
            else:
                print("False")

    def inorder(self):
        if self.left:
             self.left.inorder()

        print(self.data,end=" ")

        if self.right:
            self.right.inorder()

    def delete(self,value):
        if self.data>value:
            if self.left: 
                self.left=self.left.delete(value)

            else:
                return ("no. not present in tree")

        elif self.data<value:
            if self.right:
                self.right=self.right.delete(value)

            else:
                return ("no. not present in tree")
            
        else:
            if self.left==None and self.right==None:
                return None

            elif self.right:
                min=self.right.min()
                self.data=min
                self.right=self.right.delete(min)
                return self

            elif self.left:
                max=self.left.max()
                self.data=max
                self.left=self.left.delete(max)
                return self

        return self

    #Answer 5
    def preorder(self):
        print(self.data,end=" ")

        if self.left:
             self.left.preorder()

        if self.right:
            self.right.preorder()

    #Answer 4
    def postorder(self):
        if self.left:
             self.left.postorder()

        if self.right:
            self.right.postorder()

        print(self.data,end=" ")
     
    #Answer 1
    def min(self):
        if self.left:
            return self.left.min()

        else:
            return self.data

    #Answer 2
    def max(self):
        if self.right:
            return self.right.max()

        else:
            return self.data

    #Answer 3
    def Find_Sum(self):
        s=0
        if self.left:
             s+=self.left.Find_Sum()

        if self.right:
            s+=self.right.Find_Sum()

        s+=self.data
        
        return s

def Make_BST(list):
    root=Binary_Search_Tree(list[0])
    for i in range(1,len(list)):
        root.add(list[i])

    return root

if __name__=='__main__':
    list=[54,22,4,2,65,21,87,33,22,1,65,64,97,12,92]
    root=Make_BST(list)
    print("Is 33 present in BST ?")
    root.search(33)
    print("Is 14 present in BST ?")
    root.search(14)
    print("minimum value in tree is: ",root.min())
    print("Maximum value in tree is: ",root.max())
    print("Sum of all values in tree is: ",root.Find_Sum())
    print("Inorder traversal of list is: ",end="")
    root.inorder()
    print("\nPreorder traversal of list is: ",end="")
    root.preorder()
    print("\nPostorder traversal of list is: ",end="")
    root.postorder()
    root.delete(4)
    print("\nafter deleting 21 inorder is: ",end="")
    root.inorder()
```
